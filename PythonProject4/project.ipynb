```python
import matplotlib.pyplot as plt
```

```python
total_revenue_per_city = df.groupby('city')['fare_amount'].sum().sort_values(ascending = False)
total_revenue_per_city.plot(kind='bar')
plt.xlabel('city')
plt.ylabel('revenue')
plt.title('total revenue per city')
plt.xticks(rotation=90, ha='right')
plt.tight_layout()
plt.show()
```

```python
ride_status_distribution =df['ride_status'].value_counts()
ride_status_distribution.plot(kind='pie', autopct='%1.1f%%')

plt.xticks(rotation=45, ha='right')
plt.title('ride status distribution')
plt.tight_layout()
plt.show()
```

```python
average_fare_per_city = df.groupby('city')['fare_amount'].mean().sort_values(ascending = False)
average_fare_per_city.plot(kind='bar')
plt.xlabel('city')
plt.ylabel('average fare')
plt.title('average fare per city')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
Vehicle_type_revenue = df.groupby('vehicle_type')['fare_amount'].sum().sort_values(ascending = False)
Vehicle_type_revenue.plot(kind='bar')
plt.xlabel('vehicle_type')
plt.ylabel('revenue')
plt.title('vehicle type revenue')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
